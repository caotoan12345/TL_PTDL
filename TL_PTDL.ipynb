```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# C·∫•u h√¨nh m·∫∑c ƒë·ªãnh cho bi·ªÉu ƒë·ªì
sns.set(style="whitegrid")

# --- T√åM L·∫†I FILE D·ªÆ LI·ªÜU ---
files = os.listdir('.')
csv_file = None
for f in files:
    if f.endswith('.csv') and 'HR' in f:
        csv_file = f
        break

if csv_file:
    df = pd.read_csv(csv_file)
    df = df.rename(columns={'sales': 'department', 'Department': 'department', 'average_montly_hours': 'avg_monthly_hours'})

    print(f"‚úÖ ƒê√£ ƒë·ªçc file '{csv_file}'. ƒêang v·∫Ω l·∫°i 5 bi·ªÉu ƒë·ªì (ƒë√£ s·ª≠a l·ªói m·∫•t ch·ªØ)...")

    # --- H√åNH 1 ---
    plt.figure(figsize=(6, 6))
    df['left'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], labels=['·ªû l·∫°i', 'Ngh·ªâ vi·ªác'], startangle=90)
    plt.title('H√¨nh 4.1: T·ª∑ l·ªá Ngh·ªâ vi·ªác')
    plt.tight_layout() # S·ª≠a l·ªói m·∫•t ch·ªØ
    plt.savefig('Hinh4.1_Fixed.png', bbox_inches='tight') # L∆∞u ·∫£nh chu·∫©n
    plt.close()

    # --- H√åNH 2 ---
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df, x='satisfaction_level', hue='left', fill=True, palette='crest')
    plt.title('H√¨nh 4.2: M·ª©c ƒë·ªô h√†i l√≤ng')
    plt.tight_layout()
    plt.savefig('Hinh4.2_Fixed.png', bbox_inches='tight')
    plt.close()

    # --- H√åNH 3 ---
    plt.figure(figsize=(8, 6))
    sns.countplot(x='salary', hue='left', data=df, order=['low', 'medium', 'high'], palette='viridis')
    plt.title('H√¨nh 4.3: L∆∞∆°ng')
    plt.tight_layout()
    plt.savefig('Hinh4.3_Fixed.png', bbox_inches='tight')
    plt.close()

    # --- H√åNH 4 ---
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='number_project', y='avg_monthly_hours', hue='left', data=df, palette='Set2')
    plt.title('H√¨nh 4.4: √Åp l·ª±c c√¥ng vi·ªác')
    plt.tight_layout()
    plt.savefig('Hinh4.4_Fixed.png', bbox_inches='tight')
    plt.close()

    # --- H√åNH 5 (S·ª≠a l·ªói quan tr·ªçng nh·∫•t ·ªü ƒë√¢y) ---
    plt.figure(figsize=(10, 8)) # TƒÉng k√≠ch th∆∞·ªõc h√¨nh
    df_temp = df.copy()
    df_temp['salary_num'] = df_temp['salary'].map({'low': 1, 'medium': 2, 'high': 3})
    numeric_df = df_temp.select_dtypes(include=['number'])

    # V·∫Ω Heatmap
    sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm',
                linewidths=0.5, cbar_kws={"shrink": .8}) # Thu nh·ªè thanh m√†u l·∫°i

    plt.title('H√¨nh 4.5: Ma tr·∫≠n t∆∞∆°ng quan')
    # L·ªánh quan tr·ªçng: T·ª± ƒë·ªông cƒÉn ch·ªânh l·ªÅ ƒë·ªÉ kh√¥ng m·∫•t ch·ªØ
    plt.tight_layout()
    # L·ªánh l∆∞u ·∫£nh: ƒê·∫£m b·∫£o l∆∞u to√†n b·ªô khung h√¨nh
    plt.savefig('Hinh4.5_Fixed.png', bbox_inches='tight', dpi=300)
    plt.close()

    print("\nüéâ XONG R·ªíI! 5 ·∫£nh m·ªõi (ƒë√£ s·ª≠a l·ªói) ƒë√£ ƒë∆∞·ª£c l∆∞u.")
    print("üëâ B·∫°n h√£y t·∫£i c√°c file c√≥ ƒëu√¥i '_Fixed.png' v·ªÅ nh√© (Nh·ªõ b·∫•m n√∫t L√†m m·ªõi/Refresh tr∆∞·ªõc).")

else:
    print("‚ùå L·ªói: Kh√¥ng t√¨m th·∫•y file d·ªØ li·ªáu. B·∫°n vui l√≤ng t·∫£i l·∫°i file HR_comma_sep.csv l√™n Colab.")
```
